````python
import numpy as np
import matplotlib.pyplot as plt
````

The most common package in Python to create professional graphs is the [matplotlib package](https://matplotlib.org/stable/). In this context, we will use the `matplotlib.pyplot.subplots()` utility wrapper to generate plots. It provides additional flexibility to using a plot directly. With `subplots()` you can create multiple plots (called **axes**) within a single figure. The method returns a tuple with the figure and all the axes. 

The following is an example to make a graph for a function $f(x)$.

````python
f = lambda x: np.sin(x)**2
x_vec = np.linspace(-2,2,1000)

fig, ax = plt.subplots()
ax.plot(x_vec, f(x_vec)) 
````

````python
f = lambda x, n: x**n      # function f(x,n)=x^n
x = np.linspace(-1,1,500)  # range of argument

fig, ax = plt.subplots(figsize=(6,2), 
     nrows=1, ncols=3, sharey=True)

for j in range(3):
  ax[j].plot(x, f(x,j+2), label=f"$y=x^{j+2}$")
  ax[j].legend()
  
plt.show()
````

The `fig` contains everything outside the plot and `ax` is the field of graph. A figure can have multiple `axes` in the sense of subplots `fig, (ax1,ax2) = plt.subplots(nrows=2)`. Most methods are implemented with `ax`.

| **function**        | **type**     |
|:--------------------|:-------------|
| ax.plot(x, y)       | line chart (models)  |
| ax.scatter(x, y)    | scatter plot (experimental data) |
| ax.bar(x, height)   | bar chart (discrete data)   |
| ax.hist(data, bins) | histogram    |
| ax.imshow(image)    | image or matrix       |
| ax.fill(x,y)        | fill polygon |
| ax.fill_between(x, y1, y2) | shade area |
| ax.contour(x, y, z) | contour plot |

The graphs can be customized and styled. There are many ways to do so, the most common include setting titles, limits, annotations, and adding a legend.  

| Function                      | Type                          |
|:------------------------------|:------------------------------|
| ax.set_xlabel(label)          | x-axis label                  |
| ax.set_title(title)           | subplot title                 |
| ax.set_xlim((left, right))    | x-axis limits                 |
| ax.text(x, y, text)           | text at specified coordinates |
| ax.annotate(text, xy, xytext) | annotates a point   |
| ax.legend()                   | display legend  |
| ax.set_xticks(ticks)          | tick positions   |
| ax.set_xticklabels([])        | x axis tick labels            |
| ax.set_aspect()               | "equal", "auto" or a number    |

Adding annotations or text to a graph facilitates physical insight. It is essential to attach a specific data point `xy`, then incorporate an arrow and corresponding label text:

``` python
ax.annotate("$\sigma=1.0$", # text
    xy=(x[10],y1[10]),      # end-point of arrow
    xytext=(1.6, 0.5),      # start-point of arrow
    arrowprops=dict(
        arrowstyle="->",    # type: "fancy","<-","<->"
        linewidth=1,
        linestyle="--",     # ":","-.","-"
        connectionstyle='arc3,rad=0.1',
        color="red"))
ax.annotate("$\sigma=1.2$", xy=(x[1],y2[1]), 
    xytext=(1.6, 0.2),
    arrowprops=dict(arrowstyle="->", color="blue"))
```
