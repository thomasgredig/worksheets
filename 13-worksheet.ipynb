## Euler Method  

The simplest method to solve an *initial value problem* (IVP) for an ordinary differential equation is the *Euler method*.    The *forward Euler algorithm* is based on numerical calculus  and uses an iterative process to compute future values. The best practical example are Newton's equations of motions $dp /dt = F_{\text{net}}$. On the left side, we have the change of momentum $dp$ or the trajectory of the particle and on the right side, we have the external net force. The predicted future momentum is

$$
p(t + \Delta t) \approx p(t) + F(t,p) \Delta t
$$  

 
### Forward (Explicit) Euler Algorithm  

Given a first-order differential equation: $\dot{p} = F \left(  t, p(t) \right)$

1.  select a time step $\Delta t$ and set initial value $p_0 = p(t_0)$
2.  set $t_{n+1} = t_n+\Delta t$
3.  set $p_{n+1} = p_n + F(t_n,p_n) \Delta t$
4.  repeat step 2
 


## Second Order Differential Equations

The forward Euler algorithm only solves first-order differential equations. However, higher order differential equations can be converted into multiple first-order differential equations.

## Euler-Cromer Method {#sec-euler-cromer-method}

While the *forward Euler algorithm* is the easiest to implement, it turns out to be only conditionally stable. In other words, if the time step is not sufficiently small, non-physical divergences can occur due to rapid variations of the ODE. It fails for seemingly simple problems, such as the spring-mass oscillator or the simple pendulum.

An improvement is accomplished by introducing a small modification into the previous algorithm using a last point approximation; this approach is called the Euler-Cromer algorithm.

If the evolving time step in Euler's method is not sufficiently small, then it becomes clear that the slope at a given point may not be a good indication of the next value. This becomes especially apparent in oscillatory problems, such as a pendulum or a mass-spring system. In Newton's equations of motion, the forward Euler algorithm updates both the position and speed based on the previous values. Therefore, it overshoots slightly at the turning points. This small overshoot quickly accumulates.

| Step | Euler                               | Euler-Cromer                        |
|:-----|-------------------------------------|-------------------------------------|
| 1    | $v_n = v_{n-1} + F_{n-1} \Delta t$  | $v_n = v_{n-1} + F_{n-1}\Delta t$   |
| 2    | $x_n = x_{n-1} + v_{n-1} \Delta t$  | $x_n = x_{n-1} + v_n \Delta t$      |

## A Priori and A Posteriori Verification

In the *a posteriori* verfication method, the computational results are compared with the expected results and a numerical error is determined **after** the computation. In the *a priori* verfication method, the discretization is analyzed **before** the computation. The model's convergence is usually computed  using *O-*notation and provides an uncertainty for the results.
