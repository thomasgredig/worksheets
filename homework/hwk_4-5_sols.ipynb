22. **(P)** A projectile is launched from ground level with an initial speed $v_0$ at an angle $\theta$ above the horizontal. From kinematics, the horizontal range $R$ and time of flight $T$ are given by $R =  v_0^2  /g \; \sin(2 \theta_0)$ and $T = 2 v_0 /g \; \sin(\theta_0)$, then  **(a)** Write a function `projectileMotion(v0, th0)` that takes the initial velocity `v0` in m/s and angle `th0` in degrees. It returns a tuple `(R, T)` with the range and time of flight.  **(b)** Using list comprehension, compute the range and flight time for angles between $10^\circ$ and $80^\circ$ (step $5^\circ$), assuming $v_0 = 20 \text{ m/s}$. Output a table with the angles and corresponding range. 

```python
# Solution (a)

from math import radians
import numpy as np

def projectileMotion(v0, th0):
    g = 9.81
    R = v0**2/g*np.sin(2*radians(th0))
    T = 2*v0/g*np.sin(radians(th0))
    return (R,T)

projectileMotion(10,45)
```

```python
# Solution (b)

TH0 = [x for x in range(10,81,5)]
for th0 in TH0:
    v0=20
    R,T = projectileMotion(20, th0)
    print(f"{v0=:3.1f} {th0=:3.1f} has range {R:5.2f} m and flight time {T:3.4f} s")
```

34. **(P)** Diagonalization of a 3x3 Matrix is common in physics. Solving SchrÃ¶dinger's equation for the bound hydrogen atom, we start with a matrix $A$ that has this shape: $$ A = \begin{bmatrix} 2 & -1 & 0 \\ -1 & 2 & -1 \\ 0 & -1 & 2 \end{bmatrix}$$ This matrix appears in discrete second-derivative approximations.  **(a)**  Use `NumPy` to find its eigenvalues and eigenvectors.  **(b)**  Verify that $A = PDP^{-1}$, where $D$ is the diagonal matrix of eigenvalues and $P$ is the eigenvector matrix.   **(c)**  Discuss the significance of the eigenvalues in terms of vibrational modes.


```python
# Solution (a)

A = [[2,-1,0],[-1,2,-1],[0,-1,2]]
np.linalg.eig(A)
```

```python
# Solution (b)

evec =  np.linalg.eig(A)
D = evec[0]*np.eye(3)
P = np.linalg.eig(A)[1]
P @ D @ np.linalg.inv(P)
```
