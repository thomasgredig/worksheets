1. **(N)** Graph the function $f(x) = \sin(x)/\sqrt{x}$ in the range $0 < x < 40$. Use a red dashed line for the function $f(x)$, label both axes and add a legend. 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Solution
f = lambda x: np.sin(x)/np.sqrt(x)
x = np.linspace(1e-5,40,1000)
fig, ax = plt.subplots()
ax.plot(x, f(x),color='red',ls='--',label=r"$f(x)=\sin(x)/\sqrt{x}$")
ax.set_xlabel("x"); ax.set_ylabel("f(x)")
ax.legend()
plt.show()
```

38. Given $f(x) = x^2 - \alpha x + 5$, graph this function for three different values of $\alpha = -1,0,+1$. Use 3 subplots with shared y-axis and common x-axis values, clearly label each subplot with the corresponding $\alpha$ value.

```python
# Solution
def f(x,a=0):
  return x**2 - a*x + 5
x = np.linspace(-1,1,100)
fig,ax = plt.subplots(ncols=3, sharey=True)
for a_val in [-1,0,1]:
  ax[a_val+1].plot(x,f(x,a_val), label=f"a={a_val}")
  ax[a_val+1].legend()
plt.show()
```

47. Evaluate the function $f(x) = \sin(x)/\sqrt{x}$ for small values of $x$. Find the smallest value $x>0$ for which you get an accurate value and compare with the known value of $\lim_{x\to 0} f(x) = 0$. 

```python
# Solution
f = lambda x: np.sin(x)/np.sqrt(x)
x=1
while f(x)>0:
  prev_x=x
  x=x/2
print(f"The smallest number that can be used is {prev_x}, which gives {f(prev_x)}.")
```
