15. The number 1634 is special, since it is the sum of its digits to the forth power; i.e. $1^4 + 6^4 + 3^4 + 4^4 = 1634$, find another 4-digit number for which this is the case. 


```python
for a in range(10):
    for b in range(10):
        for c in range(10):
            for d in range(10):
                a1 = a * 1000
                b1 = b * 100
                c1 = c * 10
                d1 = d
                sum = a1 + b1 + c1 + d1
                res = a**4 + b**4 + c**4 + d**4
                
                if sum == res and sum != 0:
                    print(f'a = {a} : b = {b} : c = {c} : d = {d}')
                    break
```

16. Consider a positive integer $m_0$. If the number $m_0$ is even, then compute $m_1 = m_0 / 2$, otherwise triple it and add one $m_1 = 3m_0 + 1$. Repeat until you reach $m_n=1$, which is the *Collatz conjecture*. (a) Find the number of steps $n$ for $m_0=7515$. (Note: check that for $m_0=1$, $n=0$ and $m_0=3$, $n=7$.) (b) Find the $0 < m_0 < 10000$ with the largest $n$.

```python
m_n = [7515]

while True:
    m0 = m_n[-1]

    if m0 % 2 == 1:
        m1 = 3*m0 + 1
    else:
        m1 = m0//2

    m_n.append(m1)

    if m1 == 1:
        break

len(m_n), m_n
```
