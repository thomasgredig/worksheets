## Midpoint Rule

A simple way for doing numerical *integration* is to approximate the integral with a sum. The integral is split into $n$ equally spaced sections of width $\Delta x$ and evaluated at the mid-points $m_i$ of each section. 

$$
\int_a^b f(x) \; dx \approx \Delta x \sum_{i=1}^n f(m_i)
$$

The mid-points are found with $m_i = a + (i - \frac{1}{2}) \cdot \Delta x$ with $\Delta x = \frac{b-a}{n}$.

## Trapezoidal Rule  

A small variation and also intuitive numerical method to approximate the definite integral of a function over a given interval is the *trapezoidal method*. Suppose we want to find the integral of a function $f(x)$ from $a$ to $b$, then you can rewrite the integral as a sum.  

$$
\int_a^b f(x) \; dx  \approx \; \sum_{i=0}^{n-1} A_i
$$

## Simpson's Rule  
 
*Simpson's Rule* provides a more accurate estimate than the midpoint or trapezoidal rules. The algorithm is more complex, but automatically part of the `SciPy` package under `integrate.simpson` you can access the algorithm by providing two vectors of equal lengths, the y- and x-values of the function to integrate.   (Note: a numerical array is required, not a function as input for `simpson()`)

```{python}
f = lambda x: x**3
a, b, N = 0, 2, 10
x_vec = np.linspace(a,b,N)
simpson(f(x_vec), x_vec) 
```

### Simpson's Algorithm

1.  Divide the interval $[a, b]$ into sub-intervals (even number)
2.  Fit a quadratic polynomial through three points: $[x_{i-1}, x_i, x_{i+1}]$
3.  Integrate this quadratic polynomial exactly over each sub-interval

Therefore, we can mathematically express the integral in this form:

$$ 
\int_{a}^{b} f(x) \; dx \approx \frac{\Delta x}{3} \left[ f(x_0) + 4 \sum_{i=1}^{n/2} f(x_{2i-1}) + 2 \sum_{i=1}^{n/2-1} f(x_{2i}) + f(x_n) \right] 
$$


Previously, we could solve this equation, if we are given the initial conditions $x_0$ and $v_0$. Instead of $v_0$, we are provided with $x_f$. Therefore, we pick a non-zero $v_0=\alpha$ value arbitrarily and check whether we meet the goal. If it does not fit the boundary condition, then we repeat for a different value of $\alpha$.

## Quadrature  

The most popular way to integrate in Python is using the quadrature function `quad` from the `SciPy` package.   The idea is to find a sum of weighted function values at specific points that approximates the area under the curve. It not only integrates polynomials and exponential functions, but also piece-wise functions, special functions like the gamma function and it can handle **improper integrals with infinities** (`numpy.inf`).
