## Simple Newton-Raphson Algorithm

Given a function $f(x)$, find $x_n$ with $f(x_n)=0$.

1.  Make an initial guess $x_0$.
2.  Compute $x_{n+1}$ using the formula: $x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$
3.  Repeat this process until a sufficiently precise value is reached; i.e. $|f(x_{n+1})| < \epsilon$

## Polynomial Roots  

Specifically for polynomials, we can use `np.roots` to find solutions. It has the distinct advantage that no starting values are required.  This function returns all roots of a *polynomial* with coefficients given in an array. Suppose we have a polynomial with coefficients \[1, -20, 0, -5, 6\], which corresponds to the equation $x^4 - 20x^3 - 5x + 6 = 0$.

```{python}
coeff =        [1, -20, 0,  50,  -10] 
# polynomial: x^4 - 20x^3 + 50x - 10
print(np.roots(coeff).round(3))
```
