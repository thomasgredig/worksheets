## Symbolic Variables

Practical examples are provided to introduce the concepts of symbolic computation in Python. The `SymPy` functions are listed in the table. Sympy introduces new data types. If you type `a=5`, Python will cast `a` as type `int`, an integer.

Here is an example:

``` python 
x,z = symbols("x z")
f = x**2 - 5*x + 1
solve(f,x)
```

Here `type(f)` is sympy.core.add.Add type, which is a "sympified" object. For example, if you were to "sympify" the integer 2, you would write `sympify(2)` and the result would be a sympy.core.numbers.Integer type. 

| `SymPy`      | description                    | example            |
|:-----------|:---------------------------------|:----------------------|
| symbols()  | define symbols                          | x=symbols("x")     |
| Function() | define a function                       | f=Function("f")(x) |
| solve()    | solve an equation                       | solve(f,x)         |
| factor()   | factorize               | f.factor()         |
| expand()   | expand                                  | f.expand()         |
| simplify() | simplify a function      | f.simplify()       |
| subs()     | substitute a variable with a variable   | f.subs(x, y+1)     |
| evalf()    | substitute a variable with a number     | f.evalf(subs={x:2})|
| n()        | numerically evaluate                    | pi.n(25)           |
| latex()    | convert to LaTeX notation   | latex(1/x)         |

The most common `SymPy` functions are listed; these functions are used to integrate, differentiate, take limits, sum elements and convert an algebraic `SymPy` function into a numerical `NumPy` function for graphing with `MatPlotLib`.

| sympy                | description                  | example                 |
|:---------------------|:-----------------------------|:------------------------|
| diff(f,x)            | derivative of a function     | f.diff(x)               |

| lamdify(x,f)         | Sympy function to Numpy      | lambdify(x,f)           |
| factorial(n)         | factorial (n!)               | factorial(20)           |
| doit()               | force evaluate function      | Derivative(f,x).doit()  |
| solve()              | solve systems of equations   | solve(\[f,g\],\[x,y\])  |
