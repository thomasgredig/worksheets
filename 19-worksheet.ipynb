## Spline Fit  

When noisy data is provided, such as x-ray diffraction, we aim to seek a smooth function that does not necessarily go through all points, but shows the trend. Such functions are obtained from a spline, which applies a low-dimensional polynomial fit locally. A variety of spline methods are available. We will use the `scipy.interpolate` package with `CubicSpline` for noisy data of a continuous data set. 

The use is: `f = CubicSpline(x_data, y_data)`, so you provide the x and y data columns. It returns a function that you can call `f(x_new)` to generate new data points.
