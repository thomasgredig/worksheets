## Forward Differential Method


Numerical calculus provides us the answer in the limit of small time steps $\Delta t$:  

$$
\frac{df}{dt}  = \lim_{\Delta t \to 0} \frac{f(t+\Delta t)-f(t)}{\Delta t}
$$ 

We can now solve for the future value of $f(t+\Delta t)$ as long as we know the function's derivative. This is a *forward differential method*: 

$$
f(t+\Delta t) = f(t) +  \frac{df}{dt}  \cdot \Delta t + O(\Delta t)
$$

The error is indicated with $O(\Delta t)$; i.e. it is related to the time step $\Delta t$.  A smaller error is introduced with the *centered differential method*. 

$$
\frac{df}{dt}  = \lim_{\Delta t \to 0} \frac{f(t+\Delta t)-f(t-\Delta t)}{2\Delta t}
$$

It has an error of $O((\Delta t)^2)$ at the expense of twice the time step ($2\Delta t$). The *big O-notation* is used in the error analysis of algorithms.

## Truncation Error

One worry is the error $O(\Delta t)$ that is being introduced with each step for this computation.

## Truncation Error

The approximation of the infinite series for the differential creates a **truncation error**. It is due to a mathematical approximation. It is different from the **round-off error** we encountered earlier. 

Analyze and run the following code, which computes the error from the forward differential method and also from the centered differential method.

``` python
f = lambda x: np.sin(x) + np.exp(-x**2)
df = lambda x: np.cos(x) - 2*x*np.exp(-x**2)
x0 = -0.2
exact_diff = df(x0)

dt_vec = 2.0**(-np.arange(1, 40))
err_f = []  # forward method error
err_c = []  # centered method error

for dt in dt_vec:
    Df = (f(x0 + dt) - f(x0)) / dt
    Dc = (f(x0 + dt) - f(x0 - dt)) / (2*dt)
    err_f.append(abs(Df - exact_diff))
    err_c.append(abs(Dc - exact_diff))
```
